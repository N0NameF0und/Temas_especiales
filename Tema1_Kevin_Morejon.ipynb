```python
#Crea una función en Python que calcule el mínimo común múltiplo de dos
#números

def mcm(x, y):
#identificamos el numero mayor    
    if x > y:
        z = x
    else:
        z = y
#aumentamos en uno el numero mayor hasta que la division entre los dos numeros no exista resto 
    while(True):
        if((z % x == 0) and (z % y == 0)):
            respuesta = z
            break
        z += 1

    print("El minimo comun multiplo es "+str(respuesta))

    
mcm(5, 7)


```

```python

def mcd(x, y):
    #distingimos si son diferentes de 0
    while y != 0:
        #Usamos el algoritmo de euclides
        (x, y) = (y, x % y)
    print("El maximo comun divisor es "+str(x))    
    
mcd(3,6)
```

```python
import matplotlib.pyplot as plt
import random
first = ("Super", "Retarded", "Great", "Sexy", "Vegan", "Brave", "Shy", "Cool", "Poor", "Rich", "Fast", "Gummy", "Yummy", "Masked", "Unusual", "American", "Bisexual", "MLG", "Mlg", "lil", "Lil")
second = ("Coder", "Vegan", "Man", "Hacker", "Horse", "Bear", "Goat", "Goblin", "Learner", "Killer", "Woman", "Programmer", "Spy", "Stalker", "Spooderman", "Carrot", "Goat", "Quickscoper", "Quickscoper")

# Creamos las propiedades del grafico
labels = random.choice(first),random.choice(second),random.choice(second)
sizes = [120, 120, 120]
colors = ['yellow', 'green', 'coral']
explode = (0.3, 0, 0)  

# Iniciamos el objeto del grafico con las propiedades
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', startangle=120)

plt.axis('equal')
plt.show()
```
